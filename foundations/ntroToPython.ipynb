```python
# Try sorting this list
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
```

```python
for day in days:
    if(day == "Fri"):
        print("TGIF!")
    elif(day == "Sat" or day == "Sun"):
        print("Weekend - yay!")
    else:
        print("Weekday...")
```

```python
# Bundle the code into a function
def dayOfWeek(day):
    if(day == "Fri"):
        return("TGIF!")
    elif(day == "Sat" or day == "Sun"):
        return("Weekend - yay!")
    else:
        return("Weekday...")
```

```python
def apply_to_list(some_list, func):
    return [func(elt) for elt in some_list]

# Declare a function to apply to the list
def sqr_func(x):
    return(x**2)

# Notice a function passed as an object
apply_to_list([1, 2, 3, 4], sqr_func)
```

```python
# Skip creating new functions as
# arguments: use a lambda expression
apply_to_list([1, 2, 3, 4],
              lambda x: x**2)
```

```python
# Print only for odd entries
apply_to_list([1, 2, 3, 4],
              lambda x: x**2 if (x % 2 == 1) else None)
```

```python
from time import time

# Use a for loop to build items in a list
def list_append():
    start_time = time()
    raw_list = []
    for x in range(1, 100000):
        raw_list.append(x**2)
    print("Raw Append time: ", round(time() - start_time, 3))

# Use comprehension to build items in a list
def list_comprehension():
    start_time = time()
    comp_list = [x ** 2 for x in range(1, 100000)]
    print("Comprehension time: ", round(time() - start_time, 3))
```

```python
list_append()
list_comprehension()
```
